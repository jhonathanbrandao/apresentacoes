#### Recuperando tendências

```python
# The Yahoo! Where On Earth ID para o mundo inteiro é 1.
# Veja https://dev.twitter.com/docs/api/1.1/get/trends/place e
# http://developer.yahoo.com/geo/geoplanet/ para outras lozalicações

WORLD_WOE_ID = 1
US_WOE_ID = 23424977
BR_WOE_ID = 23424768

# Prefix ID with the underscore for query string parameterization.
# Without the underscore, the twitter package appends the ID value
# to the URL itself as a special case keyword argument.

world_trends = twitter_api.trends.place(_id=WORLD_WOE_ID)
us_trends = twitter_api.trends.place(_id=US_WOE_ID)
br_trends = twitter_api.trends.place(_id=BR_WOE_ID)

print (world_trends)
print ()
print (us_trends)
print ()
print (br_trends)
```
