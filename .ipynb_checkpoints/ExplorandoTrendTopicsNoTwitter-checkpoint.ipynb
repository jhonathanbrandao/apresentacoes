# Explorando os Trending Topics do Twitter

### Autorizando uma aplicação a acessar os dados do Twitter

```python
import twitter

# XXX: Vá em http://dev.twitter.com/apps/new para criar uma nova aplicação e pegar
# estas credenciais, tudo, que você precisa fazer é copiar os dados
# dos valores de string vazias.
# Veja https://dev.twitter.com/docs/auth/oauth para mais informações 
# da implicamentação OAuth para twitter.

CONSUMER_KEY = ''
CONSUMER_SECRET = ''
OAUTH_TOKEN = ''
OAUTH_TOKEN_SECRET = ''

auth = twitter.oauth.OAuth(OAUTH_TOKEN, OAUTH_TOKEN_SECRET,
                           CONSUMER_KEY, CONSUMER_SECRET)

twitter_api = twitter.Twitter(auth=auth)


print (twitter_api)
```

### Recuperando trends

```python
# The Yahoo! Where On Earth ID para o mundo inteiro é 1.

WORLD_WOE_ID = 1
US_WOE_ID = 23424977
BR_WOE_ID = 23424768

world_trends = twitter_api.trends.place(_id=WORLD_WOE_ID)
us_trends = twitter_api.trends.place(_id=US_WOE_ID)
br_trends = twitter_api.trends.place(_id=BR_WOE_ID)

print (world_trends)
print ()
print (us_trends)
print ()
print (br_trends)
```

### Mostrando a resposta da API em JSON

```python
import json

print (json.dumps(world_trends, indent=1))
print ()
print (json.dumps(us_trends, indent=1))
```

### Computando a interseção entre os trends mundiais e brasileiros

```python
world_trends_set = set([trend['name'] 
                        for trend in world_trends[0]['trends']])

br_trends_set = set([trend['name'] 
                     for trend in br_trends[0]['trends']]) 

common_trends = world_trends_set.intersection(br_trends_set)

print (common_trends)
```

### Coletando resultados de pesquisa

```python
# Importe unquote para evitar erros de codificação de url em next_results
from urllib.parse import unquote

q = '#DiaNacionalDoLivro' 

count = 100

# Veja https://dev.twitter.com/docs/api/1.1/get/search/tweets

search_results = twitter_api.search.tweets(q=q, count=count)

statuses = search_results['statuses']


# Iterar em 5 lotes do resultado seguindo o cursor

for _ in range(5):
    print ("Tamanho do statuses", len(statuses))
    try:
        next_results = search_results['search_metadata']['next_results']
    except (KeyError, e): 
        break
        
    # Crie um dicionário de next_results, que tem o seguinte formulário:
    # ?max_id=313519052523986943&q=NCAA&include_entities=1
    kwargs = dict([kv.split('=') for kv in unquote(next_results[1:]).split("&") ])    

    
    search_results = twitter_api.search.tweets(**kwargs)
    statuses += search_results['statuses']

# Mostrar um resultado de pesquisa fatiando a lista...
print (json.dumps(statuses[0], indent=1))
```

### Extraindo as entidades dos tweets - texto, nomes e hashtags

```python
status_texts = [ status['text'] 
                 for status in statuses ]

screen_names = [ user_mention['screen_name'] 
                 for status in statuses
                     for user_mention in status['entities']['user_mentions'] ]

hashtags = [ hashtag['text'] 
             for status in statuses
                 for hashtag in status['entities']['hashtags'] ]

# Compute a collection of all words from all tweets
words = [ w 
          for t in status_texts 
              for w in t.split() ]

# Explore the first 5 items for each...

print (json.dumps(status_texts[0:5], indent=1))
print (json.dumps(screen_names[0:5], indent=1))
print (json.dumps(hashtags[0:5], indent=1))
print (json.dumps(words[0:5], indent=1))
```

### Criando uma distribuição básica de frequência a partir das palavras nos tweets

```python
from collections import Counter

for item in [words, screen_names, hashtags]:
    c = Counter(item)
    print (c.most_common()[:10]) # top 10
    print ()
```

### Usando prettytable para exibir tuplas em um formato tabular agradável

```python
from prettytable import PrettyTable

for label, data in (('Palavras', words), 
                    ('Nome', screen_names), 
                    ('Hashtag', hashtags)):
    pt = PrettyTable(field_names=[label, 'Contador']) 
    c = Counter(data)
    [ pt.add_row(kv) for kv in c.most_common()[:10] ]
    pt.align[label], pt.align['Contador'] = 'l', 'r' # setando alinhamento da coluna
    print (pt)
```

### Calculando a diversidade lexical dos tweets

```python
# Uma função para computar a diversidade lexical

def lexical_diversity(tokens):
    return 1.0*len(set(tokens))/len(tokens)

# Uma função para calcular o número médio de palavras por tweet
def average_words(statuses):
    total_words = sum([ len(s.split()) for s in statuses ]) 
    return 1.0*total_words/len(statuses)

print (lexical_diversity(words))
print (lexical_diversity(screen_names))
print (lexical_diversity(hashtags))
print (average_words(status_texts))
```

### Plotando frequência das palavras

```python
# import de matplotlib, inline
import matplotlib.pyplot as plt
%matplotlib inline
# import numpy as np

word_counts = sorted(Counter(words).values(), reverse=True)

plt.loglog(word_counts)
plt.ylabel("Frequencia")
plt.xlabel("Rank de Palavras")
```

### Gerando histogramas de palavras, nomes e hashtags

```python
for label, data in (('Palavras', words), 
                    ('Nomes', screen_names), 
                    ('Hashtags', hashtags)):

    # Construindo um mapa de frequência para cada conjunto de dados
    # e plotando os valores
    c = Counter(data)
    plt.hist(list(c.values())) # No python 3 é preciso converter o objeto c.values para lista
    
    # Adicionando título e y-label ...
    plt.title(label)
    plt.ylabel("Número de itens")
    plt.xlabel("Quantidade de vezes que o item apareceu)")
    
    # ... desenhando como uma nova figura
    plt.figure()
```
